```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
data = df.groupby("Parental_Involvement")["Exam_Score"].mean()

categories = data.index
scores = data.values
```

```python
plt.figure()
plt.bar(categories, scores, color=["lightblue", "lightcyan", "powderblue"])
plt.title("Low Contrast Color Palette")
plt.xlabel("Parental Involvement")
plt.ylabel("Average Exam Score")
plt.show()
```

```python
plt.figure()
plt.bar(categories, scores, color=["darkblue", "orange", "green"])
plt.title("High Contrast Color Palette")
plt.xlabel("Parental Involvement")
plt.ylabel("Average Exam Score")
plt.show()
```

```python
plt.figure()
plt.bar(categories, scores, color=["#0072B2", "#D55E00", "#009E73"])
plt.title("Colorblind-Friendly Palette")
plt.xlabel("Parental Involvement")
plt.ylabel("Average Exam Score")
plt.show()
```
