```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
print(df.head())
```

```python
bar_data = df.groupby("Parental_Involvement")["Exam_Score"].mean()

plt.figure()
plt.bar(bar_data.index, bar_data.values)
plt.title("Average Exam Score by Parental Involvement")
plt.xlabel("Parental Involvement")
plt.ylabel("Average Exam Score")
plt.show()
```

```python
line_data = df.groupby("Hours_Studied")["Exam_Score"].mean()

plt.figure()
plt.plot(line_data.index, line_data.values, marker='o')
plt.title("Exam Score Trend by Study Hours")
plt.xlabel("Study Hours Per Day")
plt.ylabel("Average Exam Score")
plt.show()
```

```python
pie_data = df["Motivation_Level"].value_counts()

plt.figure()
plt.pie(pie_data.values,
        labels=pie_data.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Motivation Level Distribution")
plt.show()
```
